```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
df = pandas.read_csv('data.csv', engine='python', sep =";")
#df.set_index("mağaza", inplace= True)
# print whole sheet data
print(df)
```

```python
df.describe()
```

```python
test = df.loc[df["tarih"] >21]
train = df.loc[df["tarih"] <=20]
```

```python
X_test = pd.DataFrame(test[["tarih", "mağaza"]].values, columns= ["tarih","magaza"])
X_train = (pd.DataFrame(train[["tarih", "mağaza"]].values, columns= ["tarih","magaza"]))
y_test = (pd.DataFrame(test[["kasa adet"]].values, columns= ["kasa_adet"]))
y_train = (pd.DataFrame(train[["kasa adet"]].values, columns= ["kasa_adet2"]))
```

```python
X_test.iloc[7]
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = pd.DataFrame(regressor.predict(X_test), columns= ["kasa_adet2"])
```

```python
y_pred
```

```python
#df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}, index=[0])
df1 = pd.concat([y_test, y_pred], axis=1)
df1.rename(columns={'kasa_adet': 'Actual', "kasa_adet2": "Pred"})
```

```python
(df1.head(25)).plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
